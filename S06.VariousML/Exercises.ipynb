# Exercieses

## Read the data

```python
from sklearn.datasets import load_breast_cancer

cancer = load_breast_cancer()
```

## Train and test split

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(cancer.data, cancer.target, random_state=0)
```

## Feature scaling

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)
```

## SVM

### Build a model

```python
from sklearn.svm import SVC

svm = SVC(kernel='rbf')
svm.fit(X_train_scaled, y_train)
```

### Predict & evaluate

```python
from sklearn.metrics import accuracy_score

y_train_pred = svm.predict(X_train_scaled)
print("Accuracy on training set: {:.2f}".format(accuracy_score(y_train, y_train_pred)))

y_test_pred = svm.predict(X_test_scaled)
print("Accuracy on test set: {:.2f}".format(accuracy_score(y_test, y_test_pred)))
```

### Visualize the results

```python
import matplotlib.pyplot as plt
import numpy as np

axis_1=0
axis_2=1

x = np.concatenate((X_train_scaled, X_test_scaled))
y = np.concatenate((y_train, y_test))
y_pred = np.concatenate((y_train_pred, y_test_pred))

group_0 = x[y==0]
plt.scatter(group_0[:,axis_1], group_0[:,axis_2], c='blue', label='class 0')

group_1 = x[y==1]
plt.scatter(group_1[:,axis_1], group_1[:,axis_2], c='red', label='class 1')

plt.xlabel(cancer.feature_names[axis_1])
plt.ylabel(cancer.feature_names[axis_2])
plt.legend()
plt.show()
```

## k-means

### Build a model

```python
from sklearn.cluster import KMeans

kmeans = KMeans(n_clusters=2, random_state=0)
kmeans.fit(X_train_scaled)
```

### Predict & evaluate

```python
from sklearn.metrics import accuracy_score

y_train_pred = kmeans.predict(X_train_scaled)
print("Accuracy on training set: {:.2f}".format(accuracy_score(y_train, y_train_pred)))

y_test_pred = kmeans.predict(X_test_scaled)
print("Accuracy on test set: {:.2f}".format(accuracy_score(y_test, y_test_pred)))
```
